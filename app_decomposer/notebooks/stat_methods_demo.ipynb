#### Conclusions and future works:
- These methods are fast O(N) or O(KN), KernelCPD/Pelt -> O(CKN), BinSeg -> O(CNlogN), Botup-> O(NlogN), Wind -> O(wN)
- They do not generalize to the phases we could "see" on the signal
- We should learn on the distance(signal, reconstructured representation) -> hey but it is already the case
- It should learn the parameters on the signal it self (self-supervised learning) -> many states level (not binary) 
- Multi-level method will make phase/slice labelling easier (compute/data)
